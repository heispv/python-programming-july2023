# Matplotlib

```python
import matplotlib.pyplot as plt
```

> be komake `plt.plot(x, y)` mitoonim nemoodar x va y ro rasm konim, deghat konim ke x, y bayad numpy array e 1D bashand.

### Plot

```python
x = np.array([1, 2, 3, 4])
y = np.array([2, 6, 1, 3])

plt.plot(x, y)
```

> be komake `plt.xlabel(str)` mitoonin label x etoon ro moshakhas konid. va be komake `plt.ylabel(str)` mitoonim label y etoon ro moshakhas konid.

```python
x = np.array([1, 2, 3, 4, 7])
y = np.array([2, 6, 3, 3, 7])


plt.plot(x, y)
plt.xlabel('my X')
plt.ylabel('my Y')
plt.show()
```

```python
x = np.array([1, 2, 3, 4, 7])
y = np.array([2, 6, 3, 3, 7])


plt.plot(x, y, 'X:g')
plt.xlabel('my X')
plt.ylabel('my Y')
plt.title('My green plot')
plt.grid()
```

### Scatter Plot

```python
import random

meters_list = []
price_list = []

for i in range(100):
    meter = random.randint(60, 300)
    price = random.randint(200, 400)

    meters_list.append(meter)
    price_list.append(price)

meters = np.array(meters_list)
price = np.array(price_list)
```

```python
plt.scatter(meters, price)
plt.xlabel('Square Meters (m**2)')
plt.ylabel('Pirce (K $)')
plt.title('House price in US')
plt.grid()
plt.show()
```
